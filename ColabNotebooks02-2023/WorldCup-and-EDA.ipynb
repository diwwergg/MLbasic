```python
def appendDataName(text):

  dataName = []
  for line in text.split('\n'):
    temp = line.split('\t')[2]
    temp = temp.split(' ')
    shortname = temp[0][0] + '. ' + temp[1] 
    if len(temp) > 2 and '(' not in temp[2]: 
      shortname += ' ' + temp[2]
    print(shortname)
    dataName.append(shortname)
  return dataName
```
