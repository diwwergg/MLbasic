```python
def freqSound(A):
  # Calculate the sample rate from the length of the waveform
  sr = len(A)

  # Calculate the duration of the waveform
  T = len(A) / sr

  # Calculate the Fourier transform of the waveform
  A_fft = np.fft.fft(A)

  # Calculate the frequencies corresponding to the values in A_fft
  frequencies = np.fft.fftfreq(len(A), d=1/sr)

  # Find the index of the largest frequency in frequencies
  idx = np.argmax(np.abs(A_fft))

  # The frequency of the waveform is the largest frequency in the frequencies array
  f = frequencies[idx]

  # Return the frequency
  return abs(f)
```

```python
def closest_note(frequency, Cmaj1):
    closest_distance = float('inf')
    closest_note = None
    for note, note_frequency in Cmaj1.items():
        distance = abs(frequency - note_frequency)
        if distance < closest_distance:
            closest_distance = distance
            closest_note = note
            closest_freq = note_frequency
    print("Distance value ", closest_distance)
    return closest_note, closest_freq
```

```python
def check_note_proximity(f, Cmaj):
    closest_note = None
    closest_freq = float('inf')
    for note, freq in Cmaj.items():
        if abs(f - freq) < abs(closest_freq - f):
            closest_note = note
            closest_freq = freq
    return closest_note, closest_freq
```
