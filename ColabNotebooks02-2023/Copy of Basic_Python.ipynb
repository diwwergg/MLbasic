# Quiz

จงสร้าง Generator ที่คืนค่าจำนวนเฉพาะตัวถัดไปเรื่อย ๆ

```python
def prime():
  primes = []
  i = 2
  while True:
    flag = True
    for p in primes:
      if i % p == 0:
        flag = False
    if flag:
      primes.append(i)
      yield i
    i += 1
```

```python
x = prime()
```

```python
type(x)
```

```python
next(x)
```

answer prime number

```python
def prime():
  primes = []
  i = 2
  while True:
    flag = True
    max_divisor = math.sqrt(i)  # find the maximum divisor to check
    for p in primes:
      if p > max_divisor:  # stop checking if we've reached the maximum divisor
        break
      if i % p == 0:
        flag = False
    if flag:
      primes.append(i)
      yield i
    i += 1
```

```python
n = 10
for i, x in enumerate(prime()):
  print(i, x)
  if i+1 >= n:
    break
```

## Decorater

```python
def decorater_func(func):
  def decorated_func():
    print('before')
    func()
    print('after')
  return decorated_func

def func():
  print('this is func')

decorater_func(func)()
```

```python
def decorater_func(func):
  def decorated_func():
    print('before')
    func()
    print('after')
  return decorated_func

@decorater_func  # <------------
def func():
  print('this is func')

func() # Always decorated
```

```python
@decorater_func
def func2():
  print('this is func2')

func2()
```

```python
import time

def stop_watch(n_time):
  def inner(func):
    def wrapper(*args):
      sum_time = 0
      for i in range(n_time):
        t1 = time.time()
        func(*args)
        sum_time += time.time()-t1
      print(f'[{args[1].__name__} {n_time} times] average: {sum_time / n_time} seconds')
    return wrapper
  return inner

@stop_watch(5)
def func(n, method=sorted):
  method(list(range(int(n))))
```

```python
func(1e6, min)
```

```python
from collections import Counter

def random_walk(A, start_node, step=10):
  s = [start_node]
  for i in range(step - 1):
    next_node = np.random.choice(nodes, p=A.loc[s[-1]], size=1)
    s += list(next_node)
  return s

def prob(s):
  return {k: v/len(s) for k, v in Counter(s).items()}
```

```python
s = random_walk(A, np.random.choice(nodes), step=10000)
prob(s)
```

```python
s = random_walk(A, np.random.choice(nodes), step=50000)
prob(s)
```

## Universal Functions

ufunc is a “vectorized” wrapper for a function that takes a fixed number of specific inputs and produces a fixed number of specific outputs.

Why?

```python
A = np.random.rand(10000)
B = np.random.rand(10000)
```

Pythonic 1

```python
%%timeit
C = []
for i in range(len(A)):
  C += [A[i] * B[i]]
```

Pythonic 2

```python
%%timeit
C = []
for a, b in zip(A, B):
  C += [a * b]
```

Numpy's ufunc

```python
def mul(a, b):
  return a * b

ufunc_mul = np.frompyfunc(mul, 2, 1)
```

```python
%%timeit
C = ufunc_mul(A, B)
```

Numpy built-in function



> It is implemented in compiled C code (fastest)

# Game OOP

```python
class Human:
  def __init__(self, name, hp=100):
    self.name = name
    self.hp = hp
  def say_myname(self):
    print(self.name)
  def __repr__(self):
    return f'name: {self.name}, hp: {self.hp}'
```

```python
class Warrior(Human):
  def __init__(self, name, hp=100, atk=10, def_=5):
    super().__init__(name, hp)
    self.atk = atk
    self.def_ = def_
  def attack(self, enemy):
    damage = self.atk - enemy.def_
    if (damage > 0) : enemy.hp -= damage
  def __repr__(self):  # overidden method
    #print(type(self))
    return f'name: {self.name}, hp: {self.hp}, atk: {self.atk}, def: {self.def_}'
```

```python
class Mage(Human):
  def __init__(self, name, hp=80, atk=5,matk=12, def_=2, mdef=5):
    super().__init__(name, hp)
    self.atk = atk
    self.matk = matk
    self.def_ = def_
    self.mdef = mdef 
  def attack(self, enemy):
    damage = self.atk - enemy.def_
    if (damage > 0) : enemy.hp -= damage
  def __repr__(self):  # overidden method
    #print(type(self))
    return f'name: {self.name}, hp: {self.hp}, atk: {self.atk}, matk: {self.matk} def: {self.def_}, mdef: {self.mdef}'
```

```python
class Monk(Human):
  def __init__(self, name, hp=100):
    super().__init__(name, hp)

  def heal(self, target):
    target.hp += random.randint(5, 10)
```

```python
class Dragon(Human):
  def __init__(self, name, hp=200, atk=20):
    super().__init__(name, hp)
    self.atk = atk

    # initialize def_ and mdef with random values between 5 and 10
    self.def_ = random.randint(5, 10)
    self.mdef = random.randint(5, 10)

  def attack(self, target, atk):
    target.hp -= atk

  def breathe_fire(self, target):
    target.hp -= 20
```

```python
class Game:
  def __init__(self):
    # create a few Human, Warrior, and Dragon objects to use in the game
    self.characters = [
      Warrior('Conan', 150, 15, 8),
      Mage('Merlin', 120, 10, 15, 4, 8),
      Monk('Gandalf', 150),
      Dragon('Smaug', 200, 15, 12)
    ]

  def play(self):
    player = random.choice([c for c in self.characters if isinstance(c, Human)])
    enemy = random.choice([c for c in self.characters if isinstance(c, (Warrior, Dragon))])

    print(f'Player: {player}')
    print(f'Enemy: {enemy}')

    while player.hp > 0 and enemy.hp > 0:
      player.attack(enemy)

      if enemy.hp <= 0:
        print(f'{enemy.name} was defeated!')
        break

      # allow the enemy to attack the player
      enemy.attack(player)

      # check if the player is defeated
      if player.hp <= 0:
        print(f'{player.name} was defeated!')
        break

```

```python
Game.play
```
